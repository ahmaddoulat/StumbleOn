```python
def get_topic(article_id):
    try:
        topic_id = article_topics[article_topics['article_id'] == article_id].category_id.values[0]
        return topic_id
    except:
        return None
```

```python
new_ratings['topic_id'] = new_ratings.apply( lambda row : get_topic(row['article_id']) , axis = 1 )
```

```python
new_ratings
```

```python
user_ids = list(set(new_ratings['user_id']))
```

### SECTION 4.1: Surprise

```python
topic_surprise_per_user_avg = defaultdict(dict)
for u_id in user_ids:
    curr_user = new_ratings[new_ratings['user_id'] == u_id]
    a = curr_user.groupby('topic_id').surprise.mean()
    for item in a.iteritems():
        topic_surprise_per_user_avg[u_id][item[0]] = item[1]
        
topic_surprise_per_user_avg
```

```python
user_removed_topics
```

```python
print('user_id\tTopic\tavg_surprise')
for k, v in user_removed_topics.items():
    for item in v:
        try: print(k,'\t', item,'\t', topic_surprise_per_user_avg[k][item])
        except: continue
```

### SECTION 4.2: Interest

```python
topic_interest_per_user_avg = defaultdict(dict)
for u_id in user_ids:
    curr_user = new_ratings[new_ratings['user_id'] == u_id]
    a = curr_user.groupby('topic_id').interest.mean()
    for item in a.iteritems():
        topic_interest_per_user_avg[u_id][item[0]] = item[1]
        
topic_interest_per_user_avg
```

```python
print('user_id\tTopic\tavg_surprise')
for k, v in user_removed_topics.items():
    for item in v:
        try: print(k,'\t', item,'\t', topic_interest_per_user_avg[k][item])
        except: continue
```
